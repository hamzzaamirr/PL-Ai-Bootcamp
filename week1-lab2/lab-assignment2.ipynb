## 🔸 Step 2: Student grade evaluation using conditional statement

```python
score = int(input("Enter your score:"))
if score >= 90:
    print("Your grade is A")
elif score >= 80:
    print("Your grade is B")
elif score >= 70:
    print("Your grade is C")
elif score >= 60:
    print("Your grade is D")
else:
    print("Your grade is F")
```

## 🔸 Step 6: Python program that counts the frequency of each word in a string.

```python
text = "Python is simple yet powerful. Python is easy to learn."
text = text.lower().replace(".","").split()
word_fre = {}
for word in text:
    if word in word_fre:
        word_fre[word] += 1
    else:
        word_fre[word] = 1
print(word_fre)
```
